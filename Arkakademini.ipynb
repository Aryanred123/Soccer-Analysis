```python
#Drop irrelevent column
players_df = players_df.drop(['Best Position'], axis=1)
players_df.head(5)
```

```python
#Renaming the column names
new_column_names = {
    'Known As': 'Players Name',
    'Value(in Euro)': 'Current Marketvalue'    
}
players_df = players_df.rename(columns=new_column_names)
players_df.head(5)
```

```python
#Total number of rows and columns
players_df.shape
```

```python
#Rows with duplicate data 
duplicate_rows_players_df= players_df[players_df.duplicated()]
print("Number of duplicate rows:", duplicate_rows_players_df.shape)
```

```python
players_df.count()
```

```python
#Drop duplicates
players_df=players_df.drop_duplicates()
players_df.head()
```

```python
players_df.count()
```

```python
#Find Null Values
print(players_df.isnull().sum())
```

```python
#Comparison of Clubs based on the Market Values of their players.
Club_compare= pd.pivot_table(data=players_df, index=['Club Name'], values='Current Marketvalue', aggfunc='mean').reset_index().rename(columns={'Current Marketvalue': 'Average_Marketvalue'})
Club_compare.sort_values(by='Average_Marketvalue',ascending=False, inplace= True)
```

```python
Club_compare.reset_index(inplace=True)
```

```python
Club_compare.drop(['index'],axis=1,inplace=True)
```

```python
Club_compare
```

```python
#Comparison of Clubs based on the Overall Ratings of their players.
CLub_overall= pd.pivot_table(data=players_df, index=['Club Name'], values='Overall', aggfunc='mean').reset_index().rename(columns={'Overall': 'Average_Overall'})
CLub_overall.sort_values(by='Average_Overall',ascending=False, inplace= True)
```

```python
CLub_overall.reset_index(inplace=True)
```

```python
CLub_overall.drop(['index'],axis=1,inplace=True)
```
